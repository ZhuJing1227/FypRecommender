```python
# Imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# Read the CSV File

# df = pd.read_csv("foodData.csv")
df = pd.read_csv("newfooddata2 - Sheet1.csv")
df.head()
```

```python
# Separating the Datasets into different Dataframes

col_names = []
new_df = df

for col in df.columns:  
    x = col.replace(" ", "_")
    col_names.append(x)
    
new_df.columns = [col_names]
new_df = new_df.drop("sum", axis = 1)
new_df
```

```python
# newfooddata2 - Sheet1.csv > Top_6_most_preferred_dishes
# foodData.csv > _Top_6_most_preferred_dishes

le_data = LabelEncoder()

new_df["Gender_n"] = le_data.fit_transform(new_df["Gender"])
new_df["Age_Group_n"] = le_data.fit_transform(new_df["Age_Group"])

new_df.drop(["Gender", "Age_Group", "Top_6_most_preferred_dishes"], axis = 1, inplace = True)
new_df
```

```python
'''

Gender_n 
    Female = 0
    Male = 1
    
Age_Group_n
    14 and below = 0
    15 - 24 = 1
    25 - 54 = 2
    55 - 64 = 3
    65 and above = 4
    
1. Laksa
2. Chilli Crab
3. Char Kway Teow
4. Hainanese Chicken Rice
5. Sliced Fish Soup
6. Fish and Chips
7. Chinese Economy Rice
8. Japanese Bento
9. Curry Laksa
10. Yu Sheng
11. Duck Rice
12. Fish Head Curry
13. Bak Kut Teh
14. Hokkien Mee
15. Kway Chap
16. Roti Prata
17. Black Pepper
18. Roti John
19. Cereal Prawn
20. Beef Kway Teow
21. Katong Laksa
22. Sambal Stingray
23. Crab Bee Hoon
24. Satay

'''
food_df = new_df.drop(["UserID", "Gender_n", "Age_Group_n"], axis = 1)

gender = str(input("Enter Gender : "))
age = int(input("Enter Age : "))
food_number = int(input("\nChose one Food Number (1 - 24) :\n1. Laksa\n2. Chilli Crab\n3. Char Kway Teow\n4. Hainanese Chicken Rice\n5. Sliced Fish Soup\n6. Fish and Chips\n7. Chinese Economy Rice\n8. Japanese Bento\n9. Curry Laksa\n10. Yu Sheng\n11. Duck Rice\n12. Fish Head Curry\n13. Bak Kut Teh\n14. Hokkien Mee\n15. Kway Chap\n16. Roti Prata\n17. Black Pepper\n18. Roti John\n19. Cereal Prawn\n20. Beef Kway Teow\n21. Katong Laksa\n22. Sambal Stingray\n23. Crab Bee Hoon\n24. Satay\n"))

gendern = 0
agen = 0
    
count = 0
feature = ""

for i in food_df:
    count += 1
    i = i[0]
    if food_number == count:
        feature = i

if gender == "Male".lower() or gender == "Male".upper():
    print("MALE - 1")
    gendern = 1
    if age > 0:
        if age < 15:
            print("14 and below")
            agen = 0
        elif age >= 15 and age <= 24:
            print("15 - 24")
            agen = 1
        elif age >= 25 and age <= 54:
            print("25 - 54")
            agen = 2
        elif age >= 55 and age <= 64:
            print("55 - 64")
            agen = 3
        elif age >= 65:
            print("65 and above")
            agen = 4
    else:
        print("Invalid Age")
    
elif gender == "Female".lower() or gender == "Female".upper():
    print("FEMALE - 0")
    gendern = 0
    if age > 0:
        if age < 15:
            print("14 and below")
            agen = 0
        elif age >= 15 and age <= 24:
            print("15 - 24")
            agen = 1
        elif age >= 25 and age <= 54:
            print("25 - 54")
            agen = 2
        elif age >= 55 and age <= 64:
            print("55 - 64")
            agen = 3
        elif age >= 65:
            print("65 and above")
            agen = 4
    else:
        print("Invalid Age")
    
else:
    print("Invalid Gender")
```

```python
def nth_data(df, i : int, feature : str):
    x = df.loc[i,feature]
    return x


def similar_users(feature : str):
    similar_users = []
    similar_userids = []
    for i in new_df.T:
        for k in new_df:
            k = k[0]
            if k == feature:
                if nth_data(new_df, i, k).values == 1:
                    similar_users.append(nth_data(new_df, i, k))
    return similar_users


similar_users = similar_users(feature)


def food_pickrate(feature : str):
    temp_df = pd.DataFrame(similar_users)
    temp_list = [temp_df.T.columns]
    temp_list2 = []
    for i in temp_list:
        temp_list2.append(new_df.loc[i, feature])

    return temp_list2

def all_food_pickrates():
    temp_df = pd.DataFrame(similar_users)
    name_list = []
    percentage_list = []
    for k in food_df:
        k = k[0]
        count = 0
        for i in temp_df.T:
            if nth_data(new_df, i, k).values[0] == 1:
                count += 1
        percentage = (count / len(new_df)) * 100
        name_list.append(k)
        percentage_list.append(percentage)
            
    return name_list, percentage_list

name_list, percentage_list = all_food_pickrates()

count = 0

for i in name_list:
    if i == feature:
        print("Chosen Food.")
    print(f"Name : {name_list[count]}\nPercentage : {percentage_list[count]}\n")
    count += 1
```
